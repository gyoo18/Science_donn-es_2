```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ScienceDonnées2").getOrCreate()
rdd = spark.sparkContext.parallelize([1,2,3,4,5])
```

```python
df1 = spark.read.csv("Données/archive/Property Sales of Melbourne City.csv", header=True, inferSchema=True)
df2 = spark.read.csv("Données/archive(1)/data.csv", header=True, inferSchema=True)
df3 = spark.read.csv("Données/archive(2)/Housing.csv", header=True, inferSchema=True)
df4 = spark.read.csv("Données/archive(3)/house_price_regression_dataset.csv", header=True, inferSchema=True)
df5 = spark.read.options(header=True,delimiter=';').csv("Données/archive(4)/Homes for Sale and Real Estate.csv")
df6 = spark.read.csv("Données/archive(5)/Houses.csv", header=True, inferSchema=True)
df7 = spark.read.csv("Données/archive(6)/American_Housing_Data_20231209.csv", header=True, inferSchema=True)
df8 = spark.read.csv("Données/archive(7)/HouseListings-Top45Cities-10292023-kaggle.csv", header=True, inferSchema=True)
df9 = spark.read.csv("Données/archive(8)/properties.csv", header=True, inferSchema=True)
```

```python
df1.dropna()
df1.dropDuplicates()
df2.dropna()
df2.dropDuplicates()
df3.dropna()
df3.dropDuplicates()
df4.dropna()
df4.dropDuplicates()
df5.dropna()
df5.dropDuplicates()
df6.dropna()
df6.dropDuplicates()
df7.dropna()
df7.dropDuplicates()
df8.dropna()
df8.dropDuplicates()
df9.dropna()
df9.dropDuplicates()
```

```python
from pyspark.sql.functions import col, isnan, when, count
df1.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df1.columns]).show()
df2.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df2.columns]).show()
df3.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df3.columns]).show()
df4.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df4.columns]).show()
df5.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df5.columns]).show()
df6.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df6.columns]).show()
df7.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df7.columns]).show()
df8.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df8.columns]).show()
df9.select([count(when(isnan(c) | col (c).isNull(), c)).alias(c) for c in df9.columns]).show()
```

```python
df1 = df1.withColumnRenamed("Price","price")
df2 = df2.withColumnRenamed("Price","price")
df3 = df3.withColumnRenamed("price","price")
df4 = df4.withColumnRenamed("House_Price","price")
df5 = df5.withColumnRenamed("Price","price")
df6 = df6.withColumnRenamed("price","price")
df7 = df7.withColumnRenamed("Price","price")
df8 = df8.withColumnRenamed("Price","price")
df9 = df9.withColumnRenamed("price","price")
```

```python
dfs = [df1,df2,df3,df4,df5,df6,df7,df8,df9]
dfsall = []
for df in dfs:
    for col in df.columns:
        if col != "price" and col != "Latitude" and col != "Longitude":
            dfsall.append(df.select(col,"price"))
        if col == "Latitude":
            dfsall.append(df.select("Latitude","Longitude","price"))
for df in dfsall:
    df.show(3)
```
